##columns
1. #3 (age) 
2. #4 (sex) (1 = male; 0 = female) 
3. #9 (cp) cp: chest pain type    
-- Value 1: typical angina    
-- Value 2: atypical angina    
-- Value 3: non-anginal pain    
-- Value 4: asymptomatic 
4. #10 (trestbps)  resting blood pressure (in mm Hg on admission to the hospital) 欠損58　
5. #12 (chol) serum cholestoral in mg/dl 　欠損７+200 
6. #16 (fbs) (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 欠損82 
7. #19 (restecg) resting electrocardiographic results    欠損1 
-- Value 0: normal    
-- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)    
-- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria 
8. #32 (thalach)   maximum heart rate achieved  欠損54 
9. #38 (exang)  exercise induced angina (1 = yes; 0 = no) 欠損54 
10. #40 (oldpeak) ST depression induced by exercise relative to rest　欠損62 

    
    
11. #41 (slope)  the slope of the peak exercise ST segment    欠損119+200 (-9
-- Value 1: upsloping    
-- Value 2: flat    
-- Value 3: downsloping 
12. #44 (ca) number of major vessels (0-3) colored by flourosopy 欠損320+300
13. #51 (thal) 3 = normal; 6 = fixed defect; 7 = reversable defect 　欠損220+250
    
    

14. #58 (num) (the predicted attribute) diagnosis of heart disease (angiographic disease status)    
-- Value 0: < 50% diameter narrowing    
-- Value 1: > 50% diameter narrowing    
(in any major vessel: attributes 59 through 68 are vessels)

##complement value    
連続：各クラスの平均値補完（nan,-9,cholの0)    
カテゴリ：5に補完(nan,-9)

```python
def continuous_complement(data):
    target_group = data.groupby("target")
    columns = data.columns
    continuous_no = np.array([0,3,4,7,9])
    continuous_col = [columns[i] for i in continuous_no]
    all_class_array = []
    for j in range(len(target_group)):
        class_array = np.zeros((len(continuous_col)+1,len(target_group.get_group(j))))
        for n,name in enumerate(continuous_col):
            valeu_mean = target_group.get_group(j)[name].mean()
            fill_nan = target_group.get_group(j)[name].fillna(valeu_mean)
            if name == "chol":
                replace_9 = fill_nan.replace(-9,valeu_mean)
                class_array[n] = replace_9.replace(0,valeu_mean)
                class_array[n+1] = j
            else :
                class_array[n] = fill_nan.replace(-9,valeu_mean)
                class_array[n+1] = j
        all_class_array.append(class_array)
    #余裕あれば改善する、脳筋でかいた
    array_01 = np.hstack([all_class_array[0],all_class_array[1]])
    array_23 = np.hstack([all_class_array[2],all_class_array[3]])
    array_0_3 = np.hstack([array_01,array_23])
    array_all = np.hstack([array_0_3, all_class_array[4]])
    all_df = pd.DataFrame(array_all)
    continuous_col = continuous_col + ["target"]
    all_df.index = continuous_col
    return all_df.T

def categry_complement(data):
    target_group = data.groupby("target")
    columns = data.columns
    categry_no = np.array([1,2,5,6,8,10,11,12])
    categry_col = [columns[i] for i in categry_no]
    all_class_array = []
    for j in range(len(target_group)):
        class_array = np.zeros((len(categry_col),len(target_group.get_group(j))))
        for n,name in enumerate(categry_col):
            fill_nan = target_group.get_group(j)[name].fillna(5)
            class_array[n] = fill_nan.replace(-9,5)
        all_class_array.append(class_array)
    #余裕あれば改善する、脳筋でかいた
    array_01 = np.hstack([all_class_array[0],all_class_array[1]])
    array_23 = np.hstack([all_class_array[2],all_class_array[3]])
    array_0_3 = np.hstack([array_01,array_23])
    array_all = np.hstack([array_0_3, all_class_array[4]])
    all_df = pd.DataFrame(array_all)
    all_df.index = categry_col
    return all_df.T

def categry_continuous_conect(data):
    continuou = continuous_complement(data)
    categry = categry_complement(data)
    complement_df =  pd.concat([categry, continuou], axis=1)
    return complement_df
    #ここで細かいの補正してします　これやるとnan復活する根本改善がいる
    #target_group = data.groupby("target")
    #chol_comp = []
    #for j in range(len(target_group)):
    #    valeu_mean = target_group.get_group(j)["chol"].mean()
    #    re = target_group.get_group(j)["chol"].replace(0,valeu_mean)
    #    chol_comp.append(re)
    #return chol_comp
```

```python
def data_split(data):
    sc = StandardScaler()
    data_std = sc.fit_transform(complement_df.drop("target", axis=1))
    #data_std_df = pd.DataFrame(data_std)
    x = data_std
    y = complement_df["target"]
    X_trainval, X_test, y_trainval, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
    X_train, X_val, y_train, y_val = train_test_split(X_trainval,y_trainval,test_size=0.2, random_state=0)
    return X_trainval, X_test, X_train, X_val, y_trainval, y_test, y_train, y_val

def test_svm(X_train, X_test, y_train, y_test):
    svc = SVC(kernel='rbf',C =10, gamma=0.1)
    svc.fit(X_train,y_train)
    print("test score:{}".format(svc.score(X_test,y_test)))
    
def mlp_model():
    model = Sequential()
    model.add(layers.Dense(64, activation="relu", input_shape=(13,)))
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(5, activation="softmax"))
    model.compile(optimizer="rmsprop",
                 loss="sparse_categorical_crossentropy",
                 metrics=["accuracy"])
    return model

def kfold_cross_validation(model,X_trainval,y_trainval,epoch,bactch,n_split):
    models = KerasClassifier(build_fn=model, epochs=epoch, batch_size=bactch)
    kfold = StratifiedKFold(n_splits=n_split, shuffle=True, random_state=1)
    results = cross_val_score(models, X_trainval, y_trainval, cv=kfold)
    return results

def acc_loss_plot(models, X_train, y_train, X_val, y_val, epoch, batch):
    model = models
    history = model.fit(X_train,
                        y_train,
                        epochs=epoch,
                        batch_size=batch,
                        validation_data=(X_val,y_val),
                        verbose=0)
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    acc = history.history["acc"]
    val_acc = history.history["val_acc"]
    #loss plot
    plt.figure()
    plt.plot(loss,label="train loss")
    plt.plot(val_loss, label="validation loss")
    plt.ylabel("Loss")
    plt.xlabel("Epochs")
    plt.legend()
    #accuracy plot
    plt.figure()
    plt.plot(acc,label="train accuracy")
    plt.plot(val_acc,label="validation accuracy")
    plt.ylabel("Accuracy")
    plt.xlabel("Epochs")
    plt.legend()
    
def model_prediction(models, X_trainval, y_trainval, X_test, y_test, epoch, batch):
    model = models
    model.fit(X_trainval,
              y_trainval,
              epochs=epoch,
              batch_size=batch,
              verbose=0)
    result = model.evaluate(X_test,y_test) 
    prediction = model.predict(X_test)
    prediction_label = [np.argmax(prediction[i]) for i in range(len(prediction))]
    result_matrix = confusion_matrix(y_test,prediction_label)
    class_accurace = [(result_matrix[i][i]/result_matrix[i].sum())*1 for i in range(len(result_matrix))]
    class_array = np.array(class_accurace).reshape(5,1)
    acc_matrix = np.hstack([result_matrix,class_array])
    columns_acc = ["0","1","2","3","4","accuracy"]
    acc_df = pd.DataFrame(acc_matrix,columns=columns_acc)
    plt.figure(figsize=(10,8))
    sns.heatmap(acc_df,annot=True,cmap="Blues",cbar=False)
    plt.ylabel('Truth',fontsize=15)
    plt.xlabel('Prediction',fontsize=15)
    plt.title("Overral Accuracy:{}".format(result[1]),fontsize=18)
    plt.tick_params(labelsize=14)
```
